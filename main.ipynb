```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Admin\Videos\ASM đông\CODE ASM2.xlsx")
df
```

```python
df["Last_Name"].str.lstrip()
df["Last_Name"] = df["Last_Name"].str.lstrip("...") 
df["Last_Name"] = df["Last_Name"].str.lstrip("/")  
df["Last_Name"] = df["Last_Name"].str.rstrip("_") 
df
```

```python
df["First_Name"] = df["First_Name"].str.lstrip("...") 
df["First_Name"] = df["First_Name"].str.lstrip("/")  
df["First_Name"] = df["First_Name"].str.rstrip("_")
df["First_Name"] = df["First_Name"].str.rstrip("__")
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
df
```

```python
df['Products_Quantity'] = df['Products_Quantity'].astype(str).str.replace(r'[^0-9]', '', regex=True)
df
```

```python
df['Address'] = df['Address'].str.replace(r'[^a-zA-Z0-9\s,]', '', regex=True)
df
```
